# Process Amazon Product Review Data from Data.world

```python
import pandas as pd
```

```python
df = pd.read_csv("../data/Datafiniti_Amazon_Consumer_Reviews_of_Amazon_Products.csv") #5000 version
df.head()
```

```python
df.columns #get column names
```

pick specific columns to keep 

```python
simple_df = df[["id", "name", "categories", "reviews.doRecommend", "reviews.numHelpful", "reviews.rating", "reviews.username"]]
simple_df.head() #get only specific columns
```

explode columns to be inserted into Graph studio

```python
simple_df = simple_df.assign(categories=simple_df.categories.str.split(",")).explode('categories')
```

```python
simple_df['name'] = simple_df['name'].apply(lambda x: x.replace(',', '')) #remove commas 
```
